## 1.2 Natural Language Processing


---
1. NLP adalah bidang AI yang berfokus pada interaksi antara manusia dan komputer. NLP memungkinkan mesin untuk mengerti, menginterpretasikan, dan menggenerate bahasa manusia
2. Tujuan dari NLP adalah komputer tidak hanya mengerti satu kata saja namun juga mengerti konteks dari seluruh kata dalam kalimat
3.  beberapa masalah/tugas umum di NLP:
    1. **klasifikasi seluruh kalimat**: sentimen dari review, klasifikasi email spam atau bukan, menentukan similaritas dari 2 kalimat berbeda, dan menentukan apakah kata tersebut punya tatabahasa yang benar.
    2. **Klasifikasi tiap kata dalam kalimat**: identifikasi SPOK dalam sebuah kalimat atau nama dari objek tertentu (nama orang, lokasi, lembaga)

    4. **Ekstrak jawaban dari sebuah teks**: diberikan pertanyaan dan narasi, model akan menjawab pertanyaan berdasarkan narasi yang diberikan.
    5. **Generate kalimat baru dari teks input**: Translasi, merangkum
    6. selain data teks bisa juga buat speech recognition dan computer vision, misal subtitle atau deskripsi sebuah gambar.
4. **Kendala dalam NLP**: komputer tidak mengerti bahasa sebagaimana manusia mengerti bahasa. Teks perlu di preprocess sedemikian sehingga komputer dapat mengerti ditambah dengan kompleksnya bahasa manusia menjadi semakin sulit.

### 1.3.1 Pipeline:


Fungsi paling dasar di Transformer, pipeline menghubungkan langkah preprocessing dan postprocessing. Memungkinkan pengguna untuk memberikan input teks dan mendapat jawaban yang sesuai

**default model buat sentiment analysis** = distilbert-base-uncased-finetuned-sst-2-english

### 1.3.2 Zero Shot Classification
---
1. Klasifikasi teks yang belum di label
2. dengan pipeline zero-shot classification memungkinkan pengguna untuk menentukan label untuk klasifikasi tidak hanya positif atau negatif

**default model** = facebook/bart-large-mnli

disebut zero shot karena tidak perlu fine tune model

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

### 1.3.4 mask filling

### 1.3.5 Named Entity Recognition (NER)
---

## 1.4 Transformers, How Do They Work?

---
Keypoint:
1. ada 3 kategori transformers:
    1. GPT  (Hanya decoder, auto-regressive)
    2. BERT (Hanya encoder, auto-encoding)
    3. BART/T5 (Decoder + Encoder, sequence-to-sequence)
2. Transformers dilatih dengan **self supervised learning** -> proses pelatihan dimana fungsi objektif dihitung secara otomatis dari input model (tidak perlu label dari manusia)
3. Model yang telah dilatih memiliki pemahaman tentang bahasa manusia namun tidak tepat untuk tugas tertentu. Oleh karena itu model yang sudah dilatih melalui proses yang disebut **transfer learning** yaitu proses dimana model dilatih untuk tugas tertentu dengan dataset yang dilabel oleh manusia.
4. Transformers memerlukan data, waktu, dan komputasi yang besar untuk dilatih dari 0. Sehingga **transfer learning** menjadi hal yang penting karena dapat mengurangi kebutuhan komputasi.
5. **pretraining** adalah proses melatih model dari 0 (parameter/bobot dipilih secara random dan model belum memiliki pengetahuan apa-apa). Pretraining memerlukan data yang sangat besar dan perlu waktu berminggu minggu untuk selesai melatihnya.
6. **fine-tuning** menggunakan model **pretrained** untuk tugas tertentu. Beberapa kelebihan **fine-tuning** -> model **pretrained** sudah dilatih sebelumnya jadi memiliki pengetahuan, hal ini menyebabkan proses **fine-tuning** lebih cepat dan memerlukan data yang lebih sedikit.
7. Tranformer terdiri dari 2 bagian **encoder** (menerima input dan mengubahnya menjadi vektor representasi input) dan **decoder** (menerima hasil encoder dan input lain untuk menghasilkan sebuah kata)
8. Kunci dari Transformer adalah attention layer yang memungkin model memberikan "atensi" sebuah kata terhadap kata yang lain.
9. Arsitektur Transformer awalnya ditujukan untuk translasi, encoder menerima input dalam bahasa tertentu dan decoder menerjemahkan kalimat secara berurutan dan hanya dapat memperhatikan kata-kata yang telah diterjemahkan sebelumnya. Untuk mempercepat training, decoder diberikan seluruh kalimat target tetapi tidak diizinkan menggunakan kata-kata di masa depan.
10. Arsitektur -> kerangka dari model, penjelasan dari tiap layer yang ada dalam model. Checkpoints -> bobot yang bisa dimuat pada arsitektur. Model-> bisa arsitektur bisa juga checkpoint


## 1.5 Encoder models

---
keypoint:
1. Encoder model adalan model yang hanya menggunakan bagian encoder dari transformers.
2. Input dari encoder adalah kalimat dan akan memberikan output berupa fitur vektor (representasi dari kata tersebut)
3. fitur vektor tiap kalimat tidak hanya berisi representasi dari kata tersebut namun juga konteks kata tersebut dalam kalimat input.
4. Attention layers mampu mengakses seluruh kata dalam kalimat input. Encoder models biasa disebut sebagai bi-directional atau auto encoding models.
5. proses pretraining umumnya berupa kalimat yang tidak lengkap (ada kosongnya) dan model dilatih untuk mengisi model tersebut.
6. Encoder cocok untuk tugas yang memerlukan pemahaman dari seluruh kalimat seperti sentiment analysis, NER.
7. kalimat "my ___ name is sylvia." untuk mengisi bagian kosong ini diperlukan pemahaman dari 2 sisi (bidirectional).
8. Kelebihan dari encoder model adalah memiliki kemamuan yang baik dalam mengekstrak informasi/konteks dari sebuah kalimat
9. Beberapa contoh model ini adalah: ALBERT, BERT, DistilBERT, ELECTRA, RoBERTa

## 1.6 Decoder models

---
keypoint:
1. decoder model hany menggunakan bagian decoder dari model transformers.
2. berbeda dengan encoder, model hanya bisa mengakses kalimat yang sudah diprediksi sebelumnya, biasanya disebut dengan auto regressive.
3. pada decoder menggunakan masked self attention sehingga teks berikutnya di "tutup" dan hanya bisa berfokus pada kata sebelumnya.
3. proses pretraining umumnya berupa memprediksi kata berikutnya diberikan kata sebelumnya sebagai input.
4. Decoder cocok untuk model yang memerlukan kemampuan untuk menggenerate teks.
5. Beberapa contoh model: CTRL, GPT, GPT-2, Transformer XL


## 1.7 Sequence-to-sequence models

---
keypoint:
1. encoder-decoder model menggunakan kedua bagian dari transformers.
2. attention layer di encoder dapat menerima semua kata input sementara attention layer di decoder hanya dapat menerima kata di posisi sebelum input.
3. proses training dapat mengikuti encoder maupun decoder, namun biasanya menggunakan yang lebih kompleks. contoh, T5 dilatih dengan mengganti teks secara acak dengan kata tertentu yang ditutup dan model diminta untuk memprediksi teks yang menggantikan kata mask ini.
4.
5. sequence-to-sequence model cocok utnuk tugas yang memerlukan model menggenerate kata yang baru yang bergantung pada input, seperti merangkum, translasi atau generative question asnwering.
6. Beberapa contoh: BART, mBART, Marian, T5

## 1.8 Bias and Limitations

---
keypoint:
1. Jika ingin menggunakan pretrained atau fine-tuned untuk produksi perlu diperhatikan bahwa model juga memiliki limitasi. Biasanya adalah untuk bisa pretraining, peneliti mengambil semua konten yang tersedia baik itu bagus maupun buruk.
2. lihat code diabawah
3. Pada code dibawah terlihat kalau model bisa memiliki stigma tertentu yang didapt hasil dari pretraning. BERT bahkan dilatih tidak dengan mengambil data dari internet melainkan dari wikipedia dan BookCorpus dataset, Bahkan fine-tuning tidak dapat langsung menghapus bias yang ada sejak saat model di pretrained.

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```
