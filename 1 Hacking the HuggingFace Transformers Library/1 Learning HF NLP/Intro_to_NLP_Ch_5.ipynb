## 3.1 Slice and Dice

```python
!wget "https://archive.ics.uci.edu/ml/machine-learning-databases/00462/drugsCom_raw.zip"
!unzip drugsCom_raw.zip
```

```python
#load dataset
data_files = {"train": "drugsComTrain_raw.tsv", "test": "drugsComTest_raw.tsv"}
drug_dataset = load_dataset("csv", data_files=data_files, delimiter="\t")
```

```python
#shuffle dan ngambil 1000 sample
drug_sample = drug_dataset["train"].shuffle(seed=42).select(range(1000))
#Lihat dataset
drug_sample[:3]
```

* Unnamed: 0 adalah ID
* kolom kondisi berisi gabungan antara huruf kecil dan huruf besar
* review berisi teks yang berbeda beda panjangnya dan gabungan antara python line separator (\r\n) dan juga HTML

```python
# gunakan Dataset.unique() untuk mengecek apakah Unnamed: 0 adalah ID dengan menyamakan nomor ID dengan nomor baris
for split in drug_dataset.keys():
    assert len(drug_dataset[split]) == len(drug_dataset[split].unique("Unnamed: 0"))
```

```python
#buat lihat nama nama drug yang dipakai apa aja
drug_dataset[split].unique("drugName")
```

```python
# rename kolom Unnamed: 0 menjadi patient_id
drug_dataset = drug_dataset.rename_column(
    original_column_name="Unnamed: 0", new_column_name="patient_id"
)
drug_dataset
```

```python
# hapus None entri di dataset
drug_dataset = drug_dataset.filter(lambda x: x["condition"] is not None)
```

```python
# lowercase semua data dalam kolom condition
def lowercase_condition(example):
    return {"condition": example["condition"].lower()}


drug_dataset.map(lowercase_condition)
```

```python
# buat cek apakah sudah di lowercase semua
drug_dataset["train"]["condition"][:3]
```

## 3.2 Buat Kolom Baru

```python
# menghitung jumlah kata di tiap review
def compute_review_length(example):
    return {"review_length": len(example["review"].split())}
```

```python
drug_dataset = drug_dataset.map(compute_review_length)
# Inspect the first training example
drug_dataset["train"][0]
```

```python
# sorting dataset
drug_dataset["train"].sort("review_length")[:3]
```

```python
# hapus review yang kurang dari 30 kata karena tidak informative untuk prediksi kondisi
drug_dataset = drug_dataset.filter(lambda x: x["review_length"] > 30)
print(drug_dataset.num_rows)
```

```python
#gunakan package html untuk menghapus HTML dari sampel
import html

text = "I&#039;m a transformer called BERT"
html.unescape(text)
```

```python
drug_dataset = drug_dataset.map(lambda x: {"review": html.unescape(x["review"])})
```

```python
# proses map bsia dipercepat dengan menggunakan batched = True
drug_dataset = drug_dataset.map(lambda x: {"review": html.unescape(x["review"])}, batched = True)
```

## 3.3 Dari Dataset menjadi DataFrame dan kebalikannya

```python
#ubah ke dalam bentuk dataframe
drug_dataset.set_format("pandas")
```

```python
train_df = drug_dataset["train"][:]
drug_dataset["train"][:3]
```

```python
#ubah dari dataframe ke dalam Dataset
from datasets import Dataset

freq_dataset = Dataset.from_pandas(train_df)
freq_dataset
```

## 3.4 Buat dataset validasi

```python
#bisa dengan Dataset.train_test_split()
drug_dataset_clean = drug_dataset["train"].train_test_split(train_size=0.8, seed=42)
# Rename the default "test" split to "validation"
drug_dataset_clean["validation"] = drug_dataset_clean.pop("test")
# # Add the "test" set to our `DatasetDict`
drug_dataset_clean["test"] = drug_dataset["test"]
drug_dataset_clean
```
